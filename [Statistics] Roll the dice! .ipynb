# Roll the dice!

This time too, I want to see the effect of the law of the large number. The law of the large number state that as the size of a sample increase, the model or the probability will generally become more representative of the population from which it is drawn. 

I am using python random library to generate the outcome of the dice. The outcome would either be 1,2,3,4,5, or 6. We know that the probability of any outcome is 1/6. When we throw the dice 6 times, will any outcome of the dice "definitely" come out. The answer is yes, if we repeat the event multiple time and take the average. If we repeat the event multiple time, the probability of one outcome will be 1/6. 

To generalize this, the number of desired outcome is k. The number of the all events is n. So the probabilit is p = k / n . When the whole sample size, n is small we will see that it will not be the thoretical probability of 1/6. Why? The law of the large number. We need large sample to represent the population ie the reality.

For example, in the sample size of n = 6. There is a possibility that our desired outcome come out more than one time, which will alter the probability of the prediction. Let's say that the desired outcome come out 2 times, our probability will be 1/3 , not 1/6. But if we take large sample size, lets say 500 or 600. The probability that our desired outcome will be close to 1/6. 

```python
# import the library
import random
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# generate dice rolling event -->the outcome is [1,2,3,4,5,6]
# the outcome is random

def single_die():
    return random.randint(1,6)

single_die()
```

```python
# create a function using desired outcome and rolls count
# desired_outcome: desired outcome per one event (outcome of the dice number 1 of the 6 outcomes)
# rolls_count: the number of dice rolling 
# the function will return the number of desired outcome event happened 

def dice(desired_outcome, rolls_count):
    result = 0
    for i in range(rolls_count):
        if single_die() == desired_outcome:
            result = result + 1
    return result

# repeat the event multiple time 
# number of trials, n: number of events made (for below 1,2,3,4 ..... 1000 times)
# data: the number of desired outcome occurs, k under the number of trials, n

def search_prob(number_of_trials):
    outcome = []
    for j in range(number_of_trials):
        outcome.append(dice(1,1))
        
# for number of events n, calculate the number of desired outcome (1 in the dice) occurred : k
# then divide it with the number of events, n  to get the probability: k / n
        
    data = [x for x in outcome if x == 1]
    return (len(data)/number_of_trials)

# repeat the with different n amount
# n could be 1,2,3,4 ......1000

trials = np.arange(1,1000+1,1)
probabilities = []
for trial in trials:
    probabilities.append(search_prob(trial))

plt.figure(figsize = (15,5))
plt.plot(trials, probabilities, label = 'experiment value')
plt.plot([0,1000],[1/6,1/6], 'k-', label = 'theoretical value')
plt.xlabel('n: number of dice rolling')
plt.ylabel('p: probabilities of desired outcome')
plt.title('The Law of Large Number')
plt.legend()
plt.show()
```

Explanation: \
The graph shows the number of dice rolls and the probability of one outcome of the dice.\
The x-axis is the number of the dice rolling and y-axis is the probability of one outcome from the dice.

Probability of one outcome of the dice = number of desired outcome / total number of possible outcome

As the number of sample increase, the probability of one outcome of the dice will be close to the theoretical probability which is 1/6 or 0.16667. Note that when the number of sample is less than 100, we see lage deviation from the theoretical probability.
